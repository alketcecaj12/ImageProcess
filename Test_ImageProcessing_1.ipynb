### Explore the channels 

```python
imgRGB = cv2.imread('images/devon.jpg')
imgRGB = cv2.cvtColor(imgRGB,cv2.COLOR_BGR2RGB)

# Split the image into the B,G,R components
b,g,r = cv2.split(imgRGB)

# Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(b, cmap = 'gray');plt.title("Blue Channel");
plt.subplot(142);plt.imshow(g, cmap = 'gray');plt.title("Green Channel");
plt.subplot(143);plt.imshow(r, cmap = 'gray');plt.title("Red Channel");

```

```python
# Merge the individual channels into a BGR image
imgMerged = cv2.merge((b,g,r))
imgRGB = cv2.cvtColor(imgMerged,cv2.COLOR_BGR2RGB)

# Show the merged output
plt.imshow(imgRGB[:,:,::-1])
plt.title("Merged Output");
```

### Resize image

```python
height = imgRGB.shape[0]
width = imgRGB.shape[1]
print(width, height)
```

```python
new_height = int(imgRGB.shape[0] * 0.5)
new_width = int(imgRGB.shape[1] * 0.5)

new_size = ( new_width, new_height)

new_pic = cv2.resize(imgRGB, new_size)

cv2.imwrite('new_devon.jpg', new_pic)
```
